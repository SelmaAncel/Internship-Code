## Pearson's correlation coefficient

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

### Comparing model accuracy to engagement scores

```python
x, y = [0.58, 0.47, 0.50, 0.71, 0.68, 0.34, 0.51], [26, 25, 28, 29, 28, 23, 28]
res = stats.pearsonr(x, y)
res
```

```python
# Scatterplot
x = np.array([0.58, 0.46, 0.50, 0.71, 0.68, 0.34, 0.51])
y = np.array([26, 25, 28, 29, 28, 23, 28])

plt.scatter(x, y)
plt.show()
```
