## Chi square tests for comparing the control group to the experimental group

```python
from scipy.stats import chi2_contingency
```

### Sadness

```python
data = [[11, 9], [29, 33]]
stat, p, dof, expected = chi2_contingency(data)
print(stat)
print(p)
```

### Anger

```python
data = [[18, 21], [20, 25]]
stat, p, dof, expected = chi2_contingency(data)
print(stat)
print(p)
```

### Fear

```python
data = [[15, 9], [23, 32]]
stat, p, dof, expected = chi2_contingency(data)
print(stat)
print(p)
```
